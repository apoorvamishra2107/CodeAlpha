Correlation between features

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data = pd.read_csv('/content/Unemployment_Rate_Datsaset.csv')
data
data.info()
data.describe()
data.isnull().sum()
data.columns = ['States', 'Date', 'Frequency', 'Estimated Unemployment Rate', 'Estimated Employed',
                'Estimated Labour Participation Rate', 'Region', 'longitude', 'latitude']
data.head()

# plotting correlation heatmap
# Changed to use correct seaborn style
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(25, 30))

# Setting tick parameters
# the ax object needs to be created after the figure
fig, ax = plt.subplots()
ax.tick_params(size=20, color='w', labelsize=15, labelcolor='w')

# Computing the correlation matrix and plot the heatmap
# Selecting only numeric columns for correlation calculation
numeric_data = data.select_dtypes(include=np.number)
ax = sns.heatmap(numeric_data.corr(), annot=True, linewidth=3)

plt.show()
```

Estimated number of employee according to different region of India

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data = pd.read_csv('/content/Unemployment_Rate_Datsaset.csv')
data
data.info()
data.describe()
data.isnull().sum()
data.columns = ['States', 'Date', 'Frequency', 'Estimated Unemployment Rate', 'Estimated Employed',
                'Estimated Labour Participation Rate', 'Region', 'longitude', 'latitude']
data.head()



#Plotting histplot
data.columns = ['States', 'Date', 'Frequency', 'Estimated Unemployment Rate', 'Estimated Employed',
                'Estimated Labour Participation Rate', 'Region', 'longitude', 'latitude']
plt.title('Distribution of Estimated Unemployment Rate')
# The 'x' argument specifies the column to plot, and 'data' specifies the DataFrame.
# 'hue' is used to differentiate by region.
sns.histplot(x='Estimated Unemployment Rate', hue='Region', data=data)
plt.show()
```

Unemployment rate according to different regions of India

```python
plt.figure(figsize=(12, 6))
plt.title("Unemployment Rate by Region")
sns.histplot(x="Region", y="Estimated Unemployment Rate", data=data)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.title("Unemployment Rate by Region")
sns.barplot(x="Region", y="Estimated Unemployment Rate", data=data)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.title("Unemployment Rate by Region")
sns.histplot(x="Estimated Unemployment Rate",hue='Region',data=data)
plt.show()
```

Dashboard to analyze the unemployment rate of each state

```python
#plotting sunburst
fig = px.sunburst(data, path=['Region', 'States'], values='Estimated Unemployment Rate',
                  width=700, height=700, color_continuous_scale='RdY1Gn',
                  title="Unemployment Rate in India")
fig.show()
```
